````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('wages.csv')

# calculate the wage ratio

df['Wage Ratio'] = df['Female'] / df['Male']

# create the bar chart
ax = df.plot.bar(x="Year", y="Wage Ratio", title="Trend of wage ratio over Time")

# set the y-axis limit

plt.ylim(0,1)

# get the 3 smallest and 3 largest
df_largest3 = df.nlargest(3, "Wage Ratio")
df_smallest3 = df.nsmallest(3, "Wage Ratio")

# annotate the bars with the smallest 3 values and 3 largest values
for index, row in df_smallest3.iterrows():
    ax.text(index, row["Wage Ratio"], round(row["Wage Ratio"], 2), color = "green", ha="center")

for index, row in df_largest3.iterrows():
    ax.text(index, row["Wage Ratio"], round(row["Wage Ratio"], 2), color="red", ha="center")

# show the plot
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('wages.csv')

# calculate the wage ratio

df['Wage Ratio'] = df['Female'] / df['Male']


# create a figure and a single subplot (axes)
fig, ax = plt.subplots()

# create the bar chart
df.plot.bar(x="Year", y="Wage Ratio", title="Trend of male/female wage ratio over Time in California", ax=ax)

# set the y-axis limit

plt.ylim(0,1)

# get the 3 smallest and 3 largest
df_largest3 = df.nlargest(3, "Wage Ratio")
df_smallest3 = df.nsmallest(3, "Wage Ratio")

# annotate the bars with the smallest 3 values and 3 largest values
for index, row in df_smallest3.iterrows():
    ax.text(index, row["Wage Ratio"], round(row["Wage Ratio"], 2), color = "green", ha="center", va="bottom")

for index, row in df_largest3.iterrows():
    ax.text(index, row["Wage Ratio"] + 0.01, round(row["Wage Ratio"], 2), color="red", ha="center", va="bottom")

# add a table at the bottom
table_data = pd.concat([df_largest3, df_smallest3])[['Year', 'Wage Ratio']].values
colLabels = ["Year", "Wage Ratio"]

# plt the table baby
plt.table(cellText=table_data, colLabels = colLabels, cellLoc = 'center', loc='bottom', bbox=[0, -0.35, 1, 0.3])

# adjust the space the subplots occupy
fig.subplots_adjust(bottom=0.1)

# show the plot
plt.show()
````
